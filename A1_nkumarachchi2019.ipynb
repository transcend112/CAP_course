###"break" and "continue" statements

####### break command is important to stop a loop


```
# Prints out 0,1,2,3,4

count = 0
while True:           # run continously 
    print(count)
    count += 1
    if count >= 5:
        break         # break the loop here
        
```

```
# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:          # only output odd number , when remainder is 1
        continue            # continue the loop
    print(x)
```

image.png

###can we use "else" clause for loops?

######## else in a loop is a 'next step'. 
######## break will skip the else part , but 
######## break will not skip even if there is a continue statement 

```
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"

count=0
while(count<5):
    print(count)
    count +=1
else:
    print("count value reached %d" %(count))
```
