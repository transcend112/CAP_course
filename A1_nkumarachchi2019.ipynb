```
# Prints out 0,1,2,3,4

count = 0
while True:           # run continously 
    print(count)
    count += 1
    if count >= 5:
        break         # break the loop here
        
```

```
# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:          # only output odd number , when remainder is 1
        continue            # continue the loop
    print(x)
```

### Getting started

Numpy arrays are fast, and grate alternative to Python lists

```
# Create 2 new lists height and weight
height = [1.87,  1.87, 1.82, 1.91, 1.90, 1.85]
weight = [81.65, 97.52, 95.25, 92.98, 86.18, 88.45]

# Import the numpy package as np
import numpy as np                                  # np is the name

# Create 2 numpy arrays from height and weight
np_height = np.array(height)                    #how to create numpy arrays.
np_weight = np.array(weight)
```

### Print out the type of np_height

```
print(type(np_height))    # it will output data type , array string or whatever
```

###Element-wise calculations

Element-wise calculations; operations are very fast and computationally efficient.

```
# Calculate bmi
bmi = np_weight / np_height ** 2

# Print the result
print(bmi)
```

###Subsetting

numpy is availbe wih many subsetting, Eg; qulicky get BMI above 23

```
# For a boolean response
bmi > 23

# Print only those observations above 23
bmi[bmi > 23]
```

```
dict = {"country": ["Brazil", "Russia", "India", "China", "South Africa"],
       "capital": ["Brasilia", "Moscow", "New Dehli", "Beijing", "Pretoria"],
       "area": [8.516, 17.10, 3.286, 9.597, 1.221],
       "population": [200.4, 143.5, 1252, 1357, 52.98] }

# Set the index for brics
brics.index = ["BR", "RU", "IN", "CH", "SA"]

# Print out brics with new index values

import pandas as pd 

brics = pd.DataFrame(dict)


#############################################
brics.to_csv("cars.csv")
cars = pd.read_csv('cars.csv', index_col = 0)  # Import the cars.csv data: cars
print(cars)                                       ##### print

# Print out country column as Pandas Series
print(cars['area'])

# Print out country column as Pandas DataFrame
print(cars[['population']])

# Print out DataFrame with country and drives_right columns
print(cars[['population', 'country']])
```
